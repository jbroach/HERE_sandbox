```python
import pandas as pd
import os

drive_path = 'C:/Users/saavedrak/metro_work/HERE_sandbox/'
file = 'tmc_links_table.csv'
df_tmc_link = pd.read_csv(os.path.join(os.path.dirname('__file__'), drive_path, file))

# get list of unique tmcs:
tmcs = df_tmc_link['TMC'].drop_duplicates().values.tolist()
link_ids_by_tmc = {}
```

```python
for tmc in tmcs:
    df_temp = df_tmc_link['LINK_ID'].loc[df_tmc_link['TMC'] == tmc]
    link_ids_by_tmc[tmc] = df_temp.values.tolist()
print(link_ids_by_tmc)
```
